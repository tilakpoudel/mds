Estimate the value of pie with montecarlo method.
ie. piee = 4 * (No. of points / total no of points generated)

```python
#  Estimation value of pie with Monte Carlo Integration
import numpy as np
import matplotlib.pyplot as plt

# Method to generate random numbers using biult-in RNG
def built_in_rng(seed, n):
    np.random.seed(seed)

    return np.random.rand(n)

# Method to count number of points inside the unit circle
def count_points_inside_circle(x, y):
    return np.sum(x**2 + y**2 <= 1)

# Estimate the value of pi using Monte Carlo Integration
def estimate_pi(x, y):
    number_of_points_inside_circle = count_points_inside_circle(x, y)
    total_number_of_points = len(x)

    return (number_of_points_inside_circle / total_number_of_points) * 4

# Generate random numbers for x using built-in RNG
total_number_of_random_numbers = 100000
seed = 58942115
x = built_in_rng(seed=seed, n=total_number_of_random_numbers)
y = built_in_rng(seed=seed+1, n=total_number_of_random_numbers)

# Estimate the value of pi
pi_estimate = estimate_pi(x, y)
print(f"Estimated value of pi using Monte Carlo Integration: {pi_estimate:.8f}")

# Estimate the error
error = abs(np.pi - pi_estimate)
print(f"Error in estimation: {error:.8f}")

# Plot the error convergence with the value of number of random points
points = np.arange(500, total_number_of_random_numbers + 1, 500)
# print(points)
errors = []
pi_estimates = []

for n in points:
    x_n = built_in_rng(seed=seed, n=n)
    y_n = built_in_rng(seed=seed+1, n=n)
    count_inside_circle = count_points_inside_circle(x_n, y_n)
    pi_n = (count_inside_circle / n) * 4
    errors.append(abs(np.pi - pi_n))
    pi_estimates.append(pi_n)

# Plot the value of pi estimation convergence with the value of number of random points
plt.figure(figsize=(10, 6))
plt.plot(points, pi_estimates, label='Estimated Pi', color='blue')
plt.axhline(y=np.pi, color='red', linestyle='--', label='Actual Pi')
plt.xscale('log')
plt.xlabel('Number of Random Points')
plt.ylabel('Estimated Pi Value')
plt.title('Convergence of Pi Estimation with Monte Carlo Integration')
plt.legend()
plt.grid(True)
plt.show()

# Plot the error convergence with the value of number of random points
plt.figure(figsize=(10, 6))
plt.plot(points, errors, label='Error in Pi Estimation', color='orange')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Number of Random Points')
plt.ylabel('Error')
plt.title('Error Convergence of Pi Estimation with Monte Carlo Integration')
plt.legend()
plt.grid(True)
plt.show()

#  Plot the points and the unit circle
# plt.figure(figsize=(6,6))
# plt.scatter(x, y, s=1, color='blue', label='Random Points')
# circle = plt.Circle((0, 0), 1, color='red', fill=False, label='Unit Circle')
# plt.gca().add_artist(circle)
# plt.xlim(0, 1)
# plt.ylim(0, 1)
# plt.gca().set_aspect('equal', adjustable='box')
# plt.title('Monte Carlo Integration for Pi Estimation')
# plt.xlabel('x')
# plt.ylabel('y')
# plt.legend()
# plt.show()
```

```python
# Interpretation of results
# The estimated value of pi is close to the actual value, and the error is relatively small. 
# As we increase the number of random points, the estimation of pi converges towards the actual value, and the error decreases.
# The plots show that as we increase the number of random points, the estimated value of pi gets closer to the actual value,
# and the error decreases significantly, demonstrating the effectiveness of Monte Carlo Integration for estimating pi.
```

```python
# Estimate the value of pi 
# 1. Geometric interpretation: The ratio of the area of the unit circle to the area of the unit square is pi/4. By generating random points in the unit square and counting how many fall inside the unit circle, we can estimate pi.
# 2. Statistical interpretation: The estimation of pi can be seen as a statistical problem where we are trying to estimate the probability of a random point falling inside the unit circle. The more points we generate, the closer our estimate will be to the true value of pi, due to the Law of Large Numbers.
```
